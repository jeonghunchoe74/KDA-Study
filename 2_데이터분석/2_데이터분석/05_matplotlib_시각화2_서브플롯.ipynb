## 앤스콤 데이터셋

```python
# --------------------------
# 앤스콤 데이터셋
#   - 1973년 통계학자 프랜시스 앤스콤이 만든 데이터셋
#   - 시각화의 중요성을 보여주기 위해 설계됨
#   - 각 데이터셋은 거의 동일한 기본 통계값(갯수, 평균, 표준편차)을 가지고 있지만
#     시각화했을 때는 완전히 다른 패턴을 보여줌
# --------------------------

import seaborn as sns
anscombe = sns.load_dataset('anscombe')

df1 = anscombe[anscombe['dataset']=='I']
df2 = anscombe[anscombe['dataset']=='II']
df3 = anscombe[anscombe['dataset']=='III']
df4 = anscombe[anscombe['dataset']=='IV']
df1.head()
```

```python
display(df1.head())
display(df2.head())
display(df3.head())
display(df4.head())
```

```python
# --------------------------
# 앤스콤 데이터셋 통계수치 확인
# --------------------------

print(f'df1 >>> mean: {df1['y'].mean().round(2)}, std: {df1['y'].std().round(2)}')
print(f'df2 >>> mean: {df2['y'].mean().round(2)}, std: {df2['y'].std().round(2)}')
print(f'df3 >>> mean: {df3['y'].mean().round(2)}, std: {df3['y'].std().round(2)}')
print(f'df4 >>> mean: {df4['y'].mean().round(2)}, std: {df4['y'].std().round(2)}')
```

## 서브플롯  - 방식2
* `fig, axes = plt.subplots(nrows, ncols)`
>* 주요 옵션
>    * `figsize` : figure의 크기
>    * `shareX` : x축 눈금 공유
>    * `shareY` : y축 눈금 공유
<br>
>* 그래프 작성
>   * `axes[행][열].plot()`
>   * `axes[행][열].set_title()`
>   * `axes[행][열].set_xlabel()`
>   * `axes[행][열].set_xlim()`
<br>
>* 그래프 설정
